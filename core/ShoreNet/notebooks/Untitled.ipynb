```python
import numpy as np

# Define the arrays
a1 = np.array([1, 2, 3, 4, 3, 2, 1, 0, 0, 0])
a2 = np.array([0, 0, 0, 0.1, 0.2, 0.3, 0.4, 0.3, 0.2, 0.1])

# Calculate the correlation coefficient
correlation = np.corrcoef(a1, a2)[0, 1]

# Find the indices of the values of interest
index_4_a1 = np.where(a1 == 4)[0][0]
index_04_a2 = np.where(a2 == 0.4)[0][0]

# Output the results
print(f"Correlation coefficient: {correlation}")
print(f"Index of 4 in a1: {index_4_a1}")
print(f"Index of 0.4 in a2: {index_04_a2}")
```

```python
import matplotlib.pyplot as plt

# Plot the arrays
plt.figure(figsize=(10, 5))
plt.plot(a1, label='a1')
plt.plot(a2, label='a2')
plt.scatter(index_4_a1, a1[index_4_a1], color='red', zorder=5)
plt.scatter(index_04_a2, a2[index_04_a2], color='blue', zorder=5)
plt.legend()
plt.title('Array Comparison')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
